```
%%writefile app.py
# นี่คือโค้ด Dashboard ที่จะเก็บใน GitHub

import pandas as pd
import dash
from dash import dcc, html
import plotly.express as px
from dash.dependencies import Input, Output

# โหลดข้อมูลจาก GitHub
file_url = "https://raw.githubusercontent.com/USERNAME/REPOSITORY/main/accident2023.csv"
data = pd.read_csv(file_url)

# สร้างแอป Dash
app = dash.Dash(__name__)

# Layout ของ Dashboard
app.layout = html.Div([
    html.H1("📊 Dashboard อุบัติเหตุทางถนน 2023", style={'textAlign': 'center'}),

    dcc.Dropdown(
        id='accident_type',
        options=[{'label': i, 'value': i} for i in data['Accident_Type'].unique()],
        value=data['Accident_Type'].unique()[0],
        clearable=False
    ),

    dcc.Graph(id='accident_trend'),
])

@app.callback(
    Output('accident_trend', 'figure'),
    [Input('accident_type', 'value')]
)
def update_graph(selected_type):
    df_filtered = data[data['Accident_Type'] == selected_type]
    fig = px.bar(df_filtered, x='Month', y='Accident_Count', title="แนวโน้มอุบัติเหตุในแต่ละเดือน")
    return fig

if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
```
